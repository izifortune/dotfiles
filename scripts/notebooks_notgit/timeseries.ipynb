```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
plt.style.use('ggplot')
```

```python
df = pd.read_csv('~/Downloads/2520df8f-2afd-4a3a-ae4e-52a4fd1001aa.csv')
```

```python
df['datetime'] = pd.to_datetime(df['date'] + ' ' + df['time'], format="%Y-%m-%d %H:%M:%S")
```

```python
bins = pd.cut(df.datetime, bins=24)
```

```python
grouped = df.groupby([bins, 'status'])
```

```python
s = grouped.size()
```

```python
plot = s.unstack().plot(figsize=(15,4))
plot.xaxis.set_major_locator(mdates.HourLocator(interval=100))
plot.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
```
