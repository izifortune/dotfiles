```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
import re
plt.style.use('ggplot')
```

```python
df = pd.read_csv('~/Downloads/ee61b3ff-cd50-4e90-90a7-a39751483eb7.csv')
```

```python
df['datetime'] = pd.to_datetime(df['date'] + ' ' + df['time'], format="%Y-%m-%d %H:%M:%S")
```

```python
bins = pd.cut(df.datetime, bins=24)
```

## Top 10 Referrer

```python
# TODO Define a market list or regex to distinguish the traffic from homepageß
def group_referrer(ref):
    if "google" in ref:
        return 'google'
    elif "https://www.ryanair.com" in ref:
        if ref.endswith('/search/flights'):
            return 'homepage-mobile'
        elif "/trip/flights/" in ref:
            return ref.split('/trip/flights/')[1].split('?')[0]
        elif "ADT" in ref:
            return 'ryanair-redirect'
        else:
            return 'ryanair'
    elif "skyscanner" in ref:
        return 'skyscanner'
    elif "facebook" in ref:
        return 'facebook'
    else:
        return ref

df['ref'] = df['referrer'].apply(group_referrer)
funs = df['ref'].value_counts()
funs.sort_values(ascending=False).to_frame().iloc[0:10].plot(kind="bar")
#funs.iloc[0:10].plot(kind="pie", figsize=(15,15))
```

# Top 10 Pageview by country timeseries

```python
top10 = df['uri'].value_counts()
top10Index = top10.nlargest(10).index

filtered = df[df['uri'].isin(top10Index)]
filtered.set_index('datetime')
filtered
binstop10 = pd.cut(filtered.datetime, bins=24)

s = filtered.groupby([binstop10, 'uri']).size().sort_values(ascending=True)
plot = s.unstack().plot(figsize=(15,4))
plot.xaxis.set_major_locator(mdates.HourLocator(interval=200))
plot.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
```
