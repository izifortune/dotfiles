```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
plt.style.use('fivethirtyeight')
```

```python
dateRange = '2021-01-11-2021-01-18'
folder = '~/code/knowledge/content/ryanair/cloud-reports/data/stash-' + dateRange

data = {
'payment': pd.read_csv(folder + '/payment.ryanair.com.' + dateRange + '.csv', parse_dates=['CreatedDate']),
'homepage': pd.read_csv(folder + '/homepage.ryanair.com.' + dateRange + '.csv', parse_dates=['CreatedDate']),
'essential': pd.read_csv(folder + '/flights-essentials.ryanair.com.' + dateRange +  '.csv', parse_dates=['CreatedDate']),
'rooms': pd.read_csv(folder + '/ng2rooms.ryanair.com.' + dateRange + '.csv', parse_dates=['CreatedDate']),
'pillars': pd.read_csv(folder + '/pillar-iframes.ryanair.com.' + dateRange + '.csv', parse_dates=['CreatedDate']),
'seats': pd.read_csv(folder + '/seats.ryanair.com.' + dateRange + '.csv', parse_dates=['CreatedDate']),
'common': pd.read_csv(folder + '/common.ryanair.com.' + dateRange + '.csv', parse_dates=['CreatedDate']),
'cabinbags': pd.read_csv(folder + '/cabin-bags.ryanair.com.' + dateRange + '.csv', parse_dates=['CreatedDate']),
'checkin': pd.read_csv(folder + '/checkin.ryanair.com.' + dateRange + '.csv', parse_dates=['CreatedDate']),
'flightselect': pd.read_csv(folder + '/flights.ryanair.com.' + dateRange + '.csv', parse_dates=['CreatedDate']),
'farefinder': pd.read_csv(folder + '/ffresults.ryanair.com.' + dateRange + '.csv', parse_dates=['CreatedDate']),
'tov': pd.read_csv(folder + '/trip-overview.ryanair.com.' + dateRange + '.csv', parse_dates=['CreatedDate']),
}
```
