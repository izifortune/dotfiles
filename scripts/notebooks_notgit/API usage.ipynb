```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
plt.style.use('ggplot')

df = pd.read_csv('~/Downloads/catalog-api-referrer-2.csv')
```

```python
df['date'] = pd.to_datetime(df['day'], format='%d')

fig, ax = plt.subplots(figsize=(8,6))
for k in ['bags', 'seats', '-', 'payment', 'extras', 'checkin bags', 'checkin extras']:
    #dftotal = dftotal.append(data[k])
    df.loc[df['ref'] == k].sort_values(by="day").plot(ax=ax, x="day", y="total", label=k, figsize=(10, 5))
    
```

```python
df.groupby(['ref']).sum().sort_values(by='total', ascending=False).plot(kind='bar')
```
