## Number of pages in en-ie

```python
# Number of english pages
df = df.dropna()
df = df.dropna(subset=['cd4'])
en = df[df['cd4'].str.contains('en-ie')]
en = en.drop_duplicates(subset=['cd1'])
it = df[df['cd4'].str.contains('it-it')]
it = it.drop_duplicates(subset=['cd1'])

en.shape
```

## Pages by category

```python
# Page categories
def grouppages(x):
    return x.split(':')[0]
en['category'] = en['cd1'].apply(grouppages)
it['category'] = it['cd1'].apply(grouppages)

en['category'].value_counts().reset_index()
```

## Tagging scope

```python
def scope(x):
    if x == 'flights':
        return 'SEO pages'
    elif x == 'travel expert':
        return 'TSN'
    else:
        return '2.0'
en['category_group'] = en['category'].apply(scope)
it['category_group'] = it['category'].apply(scope)

en['category_group'].value_counts().reset_index()
```

## Type of 2.0 pages
