```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.style as styles
import matplotlib.dates as mdates
plt.style.use('seaborn-whitegrid')
```

```python
df = pd.read_csv('~/Downloads/journal.csv');
df
```

```python
df['Issues per year'] = pd.to_numeric(df['Issues per year'], errors='coerce')
df.fillna(value=1,inplace=True)
df['Issues per year'] = df['Issues per year'].astype('int')
```

```python
all = df.drop([41], axis=0)
```

```python
ax = all.plot(
    fontsize=16,
    kind='scatter', 
    x='Impact factor',
    y='Publications 2020',
    s=all["Issues per year"]*15,
    c="#3488b9",
    logx=True, logy=True, figsize=(20, 10)
)
single = df.iloc[[41]];
single.plot(ax=ax,
    fontsize=16,
    kind='scatter', 
    x='Impact factor',
    y='Publications 2020',
    s=single["Issues per year"]*15,
    logx=True, logy=True, figsize=(20, 10),
                 c='#b4252a')


import matplotlib.ticker as ticker

ax.get_xaxis().set_major_formatter(ticker.ScalarFormatter())
ax.get_yaxis().set_major_formatter(ticker.ScalarFormatter())


ax.set_title('Medicine, General & Internal Q1',pad=20, fontdict={'fontsize':24})
ax.set_ylabel('Number of Publications',fontdict={'fontsize':20})
ax.set_xlabel('Impact Factor',fontdict={'fontsize':20})

ax.set_xticks([5, 10, 30, 80])
ax.set_yticks([100, 1000, 5000])
```
